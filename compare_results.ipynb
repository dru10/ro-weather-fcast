```python
import pandas as pd
```

```python
df = pd.read_csv('results.csv')
```

```python
df
```

```python
consider = 25
```

```python
top_rmse = df.sort_values('RMSE')[:consider]
top_mape = df.sort_values('MAPE')[:consider]
top_r = df.sort_values('R', ascending=False)[:consider]
```

```python
top_rmse
```

```python
top_mape
```

```python
top_r
```

```python
promising = set.intersection(set(top_rmse.index), set(top_mape.index), set(top_r.index))
```

```python
promising_df = df.iloc[list(promising)]
promising_df
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

n_plots = len(promising_df)
plot_idx = 0

plt.figure(figsize=(16, 40))
for index, row in promising_df.iterrows():
    plot_idx += 1
    model_path = os.path.join('models', row['ModelType'], 'lags', str(row['Lags']), 'hidden', str(row['Hidden']), 'layers', str(row['Layers']), 'dropout', str(row['Dropout']), 'learning_rate', str(row["Learning_Rate"]), 'epochs', str(row['Epochs']))

    loss = mpimg.imread(os.path.join(model_path, 'loss_curves.jpg'))
    plt.subplot(n_plots, 2, plot_idx)
    plt.imshow(loss)

    plot_idx += 1

    perf = mpimg.imread(os.path.join(model_path, 'validation_performance.jpg'))
    plt.subplot(n_plots, 2, plot_idx)
    plt.imshow(perf)

    print(index)
```
