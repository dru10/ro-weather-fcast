## Imports

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

## Read csv and set index

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
# Set index as the date
index = pd.to_datetime(df[['Year', 'Month', 'Day']])
df.set_index(index, inplace=True)
df
```

```python
df.dtypes
```

## Clean columns and set correct datatypes

```python
df.index
```

```python
df['Sea_Level_Pressure'].fillna('0.0', inplace=True)
df['Sea_Level_Pressure'].unique()
```

```python
corrupted = df[df['Sea_Level_Pressure'] == 'True']
corrupted
```

```python
# Drop these rows as they are corrupted
df.drop(index=corrupted.index, inplace=True)
```

```python
df
```

### Set continuous columns to float

```python
cat_cols = [col for col in df.columns if col.startswith("Is")]
cont_cols = list(set(df.columns) - set(cat_cols) - set(["Year", "Month", "Day"]))

df[cont_cols] = df[cont_cols].astype('float')
```

```python
df
```

```python
df.dtypes
```

## Plot the data

```python
ax = df.plot(y='Temperature_Avg', c='#5884cc', figsize=(10,4), label='Average Temperature')
ax.fill_between(x=df.index, y1='Temperature_Min', y2='Temperature_Max', data=df, color=matplotlib.colors.to_rgba('#0c56cc', 0.15))
plt.legend()
plt.grid()
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.ylim((-30, 50))
plt.title("Average Temperature Bucharest Otopeni Jan 2017 - Dec 2022")
plt.savefig('plots/avg_temp.jpg')
```
